```python
from langchain_community.document_loaders import TextLoader
from langchain_core.documents import Document
import json

base = "/content/drive/MyDrive/Colab Notebooks"
scraped_dir = os.path.join(base, "IUB_Scraped_Data")
pdf_text_dir = os.path.join(base, "IUB_PDF_Text")
qa_dir = os.path.join(base, "IUB_Synthetic_QA")

all_documents = []
failed = 0

print("Loading web scraped files...")
for filename in os.listdir(scraped_dir):
    if filename.endswith(".md"):
        try:
            loader = TextLoader(os.path.join(scraped_dir, filename), encoding="utf-8")
            all_documents.extend(loader.load())
        except:
            failed += 1

print(f"After scraped files: {len(all_documents)} documents")

print("Loading PDF text files...")
for filename in os.listdir(pdf_text_dir):
    if filename.endswith(".txt"):
        try:
            loader = TextLoader(os.path.join(pdf_text_dir, filename), encoding="utf-8")
            all_documents.extend(loader.load())
        except:
            failed += 1

print(f"After PDF files: {len(all_documents)} documents")

print("Loading Q&A files...")
for filename in os.listdir(qa_dir):
    if filename.endswith(".json"):
        try:
            with open(os.path.join(qa_dir, filename), "r", encoding="utf-8") as f:
                data = json.load(f)
                if isinstance(data, list):
                    for item in data:
                        q = item.get("question", "")
                        a = item.get("answer", "")
                        if q and a:
                            all_documents.append(Document(page_content=f"Q: {q}\nA: {a}"))
        except:
            failed += 1

print(f"\nTotal documents loaded: {len(all_documents)}")
print(f"Failed: {failed}")
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

print("Splitting and filtering chunks...")

splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50,
    length_function=len
)

raw_chunks = splitter.split_documents(all_documents)
print(f"Raw chunks before filtering: {len(raw_chunks)}")

# Filter out junk chunks
junk_phrases = [
    "tenders", "careers", "downloads", "contact us",
    "my iub", "eportal", "youtube", "facebook", "twitter",
    "instagram", "linkedin", "cookie", "javascript",
    "click here", "read more", "home", "sitemap",
    "greyheader.png", "logo", "uploads/901",
    "iub.edu.pk/uploads", "[![", "](https",
]

def is_junk(chunk):
    content = chunk.page_content.lower().strip()
    # Too short to be useful
    if len(content) < 100:
        return True
    # Mostly links and navigation
    if content.count("http") > 3:
        return True
    # Contains junk phrases heavily
    junk_count = sum(1 for phrase in junk_phrases if phrase in content)
    if junk_count > 3:
        return True
    return False

clean_chunks = [c for c in raw_chunks if not is_junk(c)]

print(f"Clean chunks after filtering: {len(clean_chunks)}")
print(f"Removed {len(raw_chunks) - len(clean_chunks)} junk chunks")
print(f"\nSample clean chunk:")
print(clean_chunks[100].page_content)
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

print("Loading embedding model...")

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2",
    model_kwargs={"device": "cuda"}
)

test = embeddings.embed_query("What is IUB?")
print(f"Embedding model ready! Vector size: {len(test)}")
```

```python
import time
from langchain_chroma import Chroma

CHROMA_PATH = "/content/drive/MyDrive/Colab Notebooks/IUB_ChromaDB"
BATCH_SIZE = 1000

total_chunks = len(clean_chunks)
print(f"Total clean chunks: {total_chunks}")
print(f"Estimated batches: {total_chunks // BATCH_SIZE + 1}")
print("Starting embedding... please wait")
print("-" * 50)

start_time = time.time()

for i in range(0, total_chunks, BATCH_SIZE):
    batch = clean_chunks[i:i + BATCH_SIZE]
    batch_num = (i // BATCH_SIZE) + 1

    try:
        if i == 0:
            vectorstore = Chroma.from_documents(
                documents=batch,
                embedding=embeddings,
                persist_directory=CHROMA_PATH
            )
        else:
            vectorstore.add_documents(batch)

        elapsed = (time.time() - start_time) / 60
        remaining = (elapsed / batch_num) * ((total_chunks // BATCH_SIZE + 1) - batch_num)
        print(f"Batch {batch_num} ✅ | {i + len(batch)}/{total_chunks} chunks | Elapsed: {elapsed:.1f} min | ETA: {remaining:.1f} min")

    except Exception as e:
        print(f"Batch {batch_num} FAILED: {e}")
        time.sleep(10)
        try:
            vectorstore.add_documents(batch)
            print(f"Batch {batch_num} retry successful ✅")
        except Exception as e2:
            print(f"Skipping batch {batch_num}: {e2}")

print("-" * 50)
print(f"Done! Total chunks in ChromaDB: {vectorstore._collection.count()}")
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_chroma import Chroma
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser

# Load vectorstore
vectorstore = Chroma(
    persist_directory=CHROMA_PATH,
    embedding_function=embeddings
)

# Connect Gemini
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0.3
)

# Improved prompt
prompt_template = PromptTemplate.from_template("""You are IUB Assist, the official AI support agent for Islamia University Bahawalpur (IUB), Pakistan.
Your job is to answer questions from students, teachers, and staff about IUB.

Rules:
- Answer ONLY from the context provided below
- Prefer the most recent and specific information available
- If the context contains partial information, use it and mention it may be incomplete
- Only say "I don't have that information" if the context has absolutely nothing relevant
- Always be polite, helpful, and professional
- If the user writes in Urdu, respond in Urdu. If in English, respond in English.
- Never make up information that is not in the context

Context:
{context}

Question: {question}

Answer:""")

# Improved retriever with higher k
retriever = vectorstore.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 15}
)

def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)

rag_chain = (
    {"context": retriever | format_docs, "question": RunnablePassthrough()}
    | prompt_template
    | llm
    | StrOutputParser()
)

print("RAG pipeline ready!")
```

```python
import os
from langchain_community.document_loaders import TextLoader
from langchain_core.documents import Document
from langchain_text_splitters import RecursiveCharacterTextSplitter
import json

base = "/content/drive/MyDrive/Colab Notebooks"
scraped_dir = os.path.join(base, "IUB_Scraped_Data")
pdf_text_dir = os.path.join(base, "IUB_PDF_Text")
qa_dir = os.path.join(base, "IUB_Synthetic_QA")

all_documents = []

print("Loading web scraped files...")
for filename in os.listdir(scraped_dir):
    if filename.endswith(".md"):
        try:
            loader = TextLoader(os.path.join(scraped_dir, filename), encoding="utf-8")
            all_documents.extend(loader.load())
        except:
            pass

print(f"After scraped: {len(all_documents)}")

print("Loading PDF text files...")
for filename in os.listdir(pdf_text_dir):
    if filename.endswith(".txt"):
        try:
            loader = TextLoader(os.path.join(pdf_text_dir, filename), encoding="utf-8")
            all_documents.extend(loader.load())
        except:
            pass

print(f"After PDFs: {len(all_documents)}")

print("Loading Q&A files...")
for filename in os.listdir(qa_dir):
    if filename.endswith(".json"):
        try:
            with open(os.path.join(qa_dir, filename), "r", encoding="utf-8") as f:
                data = json.load(f)
                if isinstance(data, list):
                    for item in data:
                        q = item.get("question", "")
                        a = item.get("answer", "")
                        if q and a:
                            all_documents.append(Document(page_content=f"Q: {q}\nA: {a}"))
        except:
            pass

print(f"Total documents: {len(all_documents)}")
```

```python
from langchain_chroma import Chroma

CHROMA_PATH = "/content/drive/MyDrive/Colab Notebooks/IUB_ChromaDB"

vectorstore = Chroma(
    persist_directory=CHROMA_PATH,
    embedding_function=embeddings
)
print(f"ChromaDB loaded: {vectorstore._collection.count()} chunks")
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser

retriever = vectorstore.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 15}
)

def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)

prompt_template = PromptTemplate.from_template("""You are IUB Assist, the official AI support agent for Islamia University Bahawalpur (IUB), Pakistan.
Your job is to answer questions from students, teachers, and staff about IUB.

Rules:
- Answer ONLY from the context provided below
- Prefer the most recent and specific information available
- If the context contains partial information, use it and mention it may be incomplete
- Only say "I don't have that information" if the context has absolutely nothing relevant
- Always be polite, helpful, and professional
- If the user writes in Urdu, respond in Urdu. If in English, respond in English
- Never make up information that is not in the context

Context:
{context}

Question: {question}

Answer:""")

print("RAG pipeline ready")
```
