````python
import asyncio
import os
import time
from crawl4ai import AsyncWebCrawler, CrawlerRunConfig
from urllib.parse import urlparse

OUTPUT_DIR = "/content/drive/MyDrive/Colab Notebooks/IUB_Scraped_Data"

# IUB domains to stay within
ALLOWED_DOMAINS = {
    "iub.edu.pk",
    "www.iub.edu.pk",
    "testingservices.iub.edu.pk",
    "affiliation.iub.edu.pk",
    "library.iub.edu.pk",
    "ccsc.iub.edu.pk",
    "ssrcc.iub.edu.pk",
    "journals.iub.edu.pk",
    "dai.iub.edu.pk",
}

# Pages to skip — login-required or useless for RAG
SKIP_PATTERNS = [
    "eportal", "lms", "login", "logout", "register",
    "gallery", "video", "youtube", ".pdf", ".jpg",
    ".png", ".zip", "facebook", "twitter", "instagram",
]

already_scraped = set(os.listdir(OUTPUT_DIR))
visited_urls = set()
queue = [
    "https://iub.edu.pk/",
    "https://www.iub.edu.pk/departments",
    "https://www.iub.edu.pk/faculties",
    "https://www.iub.edu.pk/academics",
]

def url_to_filename(url):
    return (
        url.replace("https://", "")
           .replace("http://", "")
           .replace("/", "_")
           .replace("?", "_")
           .replace("=", "_")
           .replace(".", "_")
           [:80]
    ) + ".md"

def should_skip(url):
    return any(p in url.lower() for p in SKIP_PATTERNS)

def is_allowed_domain(url):
    try:
        domain = urlparse(url).netloc
        return any(domain == d or domain.endswith("." + d) for d in ALLOWED_DOMAINS)
    except:
        return False

async def deep_crawl():
    new_files = 0
    skipped = 0

    async with AsyncWebCrawler(verbose=False) as crawler:
        while queue:
            url = queue.pop(0)

            if url in visited_urls:
                continue
            visited_urls.add(url)

            filename = url_to_filename(url)
            if filename in already_scraped:
                skipped += 1
                continue

            if should_skip(url) or not is_allowed_domain(url):
                continue

            try:
                config = CrawlerRunConfig(
                    word_count_threshold=50,
                    remove_overlay_elements=True,
                    exclude_external_links=False,  # keep links for discovery
                    exclude_social_media_links=True,
                    wait_until="networkidle",
                    page_timeout=20000,
                )
                result = await crawler.arun(url=url, config=config)

                if result.success and result.markdown:
                    word_count = len(result.markdown.split())

                    # Save file
                    filepath = os.path.join(OUTPUT_DIR, filename)
                    with open(filepath, "w", encoding="utf-8") as f:
                        f.write(f"# Source URL: {url}\n")
                        f.write(f"# Scraped: {time.strftime('%Y-%m-%d')}\n\n")
                        f.write(result.markdown)
                    already_scraped.add(filename)
                    new_files += 1
                    print(f"✅ [{new_files}] {url[:65]}  →  {word_count} words")

                    # Discover new links from this page
                    if result.links:
                        for link in result.links.get("internal", []):
                            href = link.get("href", "")
                            if href and href not in visited_urls:
                                queue.append(href)

                await asyncio.sleep(1)

            except Exception as e:
                print(f"❌ {url[:55]} → {e}")

    print(f"\n✅ Deep crawl done")
    print(f"   New files saved : {new_files}")
    print(f"   Already existed : {skipped}")
    print(f"   Total in folder : {len(os.listdir(OUTPUT_DIR))}")

await deep_crawl()
````

````python
import requests
import os
import time
from urllib.parse import urlparse

PDF_DIR = "/content/drive/MyDrive/Colab Notebooks/IUB_PDFs"
os.makedirs(PDF_DIR, exist_ok=True)

PDF_LINKS_FILE = "/content/drive/MyDrive/Colab Notebooks/iub_pdf_links.txt"

with open(PDF_LINKS_FILE, "r") as f:
    pdf_urls = [line.strip() for line in f if line.strip()]

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36"
}

success = 0
failed = 0
skipped = 0
total = len(pdf_urls)

print(f"📥 Downloading {total} PDFs\n" + "─"*60)

for i, url in enumerate(pdf_urls, start=1):
    try:
        # Clean filename from URL
        filename = urlparse(url).path.split("/")[-1]
        filename = filename[:100]  # limit length
        if not filename.endswith(".pdf"):
            filename += ".pdf"

        filepath = os.path.join(PDF_DIR, filename)

        # Skip if already downloaded
        if os.path.exists(filepath):
            skipped += 1
            continue

        response = requests.get(url, headers=headers, timeout=30)

        if response.status_code == 200:
            with open(filepath, "wb") as f:
                f.write(response.content)
            size_kb = len(response.content) / 1024
            print(f"[{i}/{total}] ✅ {filename[:50]}  →  {size_kb:.0f} KB")
            success += 1
        else:
            print(f"[{i}/{total}] ❌ HTTP {response.status_code}: {url[:50]}")
            failed += 1

        time.sleep(0.5)

    except Exception as e:
        print(f"[{i}/{total}] ❌ ERROR: {url[:50]} → {e}")
        failed += 1

print("─"*60)
print(f"\n📊 Results:")
print(f"   ✅ Downloaded : {success}")
print(f"   ❌ Failed     : {failed}")
print(f"   ⏭️  Skipped    : {skipped}")
print(f"   📁 Saved to   : {PDF_DIR}")
````

````python
import re
from datetime import datetime

PDF_DIR = "/content/drive/MyDrive/Colab Notebooks/IUB_PDFs"
files = [f for f in os.listdir(PDF_DIR) if f.endswith(".pdf")]

years = {}
no_date = 0

for name in files:
    # Extract 10-digit Unix timestamp from filename
    match = re.search(r'(1[5-9]\d{8})', name)
    if match:
        ts = int(match.group(1))
        try:
            year = datetime.fromtimestamp(ts).year
            years[year] = years.get(year, 0) + 1
        except:
            no_date += 1
    else:
        # Try to find year directly in name (2015–2026)
        match2 = re.search(r'(201[5-9]|202[0-6])', name)
        if match2:
            year = int(match2.group(1))
            years[year] = years.get(year, 0) + 1
        else:
            no_date += 1

print("📅 File Age Distribution")
print("─" * 35)
for year in sorted(years):
    bar = "█" * (years[year] // 5)
    print(f"  {year} : {years[year]:>4} files  {bar}")
print("─" * 35)
print(f"  No date found : {no_date} files")
````

````python
import os
from datetime import datetime
import re

PDF_DIR = "/content/drive/MyDrive/Colab Notebooks/IUB_PDFs"

# Files to explicitly delete (too large)
DELETE_BY_NAME = [
    "annual-report-2023.pdf",
    "annual-report-2022.pdf",
]

# Keywords that make old files worth keeping regardless of year
TIMELESS_KEYWORDS = [
    "regulation", "rules", "policy", "act", "statute", "migration",
    "plagiarism", "phd", "mphil", "eligib", "prospectus", "fee-structure",
    "hostel", "scholarship", "syllabus", "course-content", "thesis", "guideline"
]

deleted = []
kept_old = []
deleted_old = []

files = [f for f in os.listdir(PDF_DIR) if f.endswith(".pdf")]

for filename in files:
    filepath = os.path.join(PDF_DIR, filename)
    name_lower = filename.lower()

    # Delete by name (giant files)
    if filename in DELETE_BY_NAME:
        os.remove(filepath)
        deleted.append(filename)
        continue

    # Check year from filename
    match = re.search(r'(1[5-9]\d{8})', filename)
    year = None
    if match:
        try:
            year = datetime.fromtimestamp(int(match.group(1))).year
        except:
            pass

    if year is None:
        match2 = re.search(r'(201[5-9]|202[0-6])', filename)
        if match2:
            year = int(match2.group(1))

    # Delete 2015-2017 files (no exceptions)
    if year and year <= 2017:
        os.remove(filepath)
        deleted.append(filename)
        continue

    # For 2018-2020: keep only timeless docs, delete the rest
    if year and year <= 2020:
        is_timeless = any(kw in name_lower for kw in TIMELESS_KEYWORDS)
        if is_timeless:
            kept_old.append(f"{year} KEPT: {filename[:60]}")
        else:
            os.remove(filepath)
            deleted_old.append(filename)
            continue

print(f"🗑️  Deleted (too large)     : {len(DELETE_BY_NAME)}")
print(f"🗑️  Deleted (2015-2017)     : {len([d for d in deleted if d not in DELETE_BY_NAME])}")
print(f"🗑️  Deleted (old 2018-2020) : {len(deleted_old)}")
print(f"✅  Kept old (timeless)     : {len(kept_old)}")
print(f"\n📁 Files remaining: {len(os.listdir(PDF_DIR))}")

print(f"\n✅ Timeless old files kept:")
for f in kept_old[:20]:
    print(f"   {f}")
````

````python
import fitz
import os
import time
import base64
import requests

PDF_DIR = "/content/drive/MyDrive/Colab Notebooks/IUB_PDFs"
PDF_TEXT_DIR = "/content/drive/MyDrive/Colab Notebooks/IUB_PDF_Text"

# Find scanned PDFs not yet extracted
extracted = set(f.replace(".txt", ".pdf") for f in os.listdir(PDF_TEXT_DIR))
all_pdfs = set(f for f in os.listdir(PDF_DIR) if f.endswith(".pdf"))
scanned_pdfs = all_pdfs - extracted

# Filter single page only
single_page = []
for filename in scanned_pdfs:
    try:
        doc = fitz.open(os.path.join(PDF_DIR, filename))
        if len(doc) == 1:
            single_page.append(filename)
        doc.close()
    except:
        pass

print(f"🎯 Remaining single-page PDFs to OCR: {len(single_page)}")

# ── OCR function with retry ───────────────────────────────
def gemini_ocr(image_bytes, retries=3):
    b64 = base64.b64encode(image_bytes).decode("utf-8")
    url = f"https://generativelanguage.googleapis.com/v1beta/models/gemini-2.5-flash:generateContent?key={GEMINI_API_KEY}"
    payload = {
        "contents": [{
            "parts": [
                {"text": "Extract ALL text from this image exactly as it appears. Preserve structure. Output only the extracted text, nothing else."},
                {"inline_data": {"mime_type": "image/png", "data": b64}}
            ]
        }]
    }

    for attempt in range(retries):
        try:
            resp = requests.post(url, json=payload, timeout=60)

            # Rate limited — wait and retry
            if resp.status_code == 429:
                wait = 30 * (attempt + 1)
                print(f"   ⏳ Rate limited. Waiting {wait}s...")
                time.sleep(wait)
                continue

            if resp.status_code != 200:
                print(f"   ❌ HTTP {resp.status_code}")
                return None

            result = resp.json()

            # Check for empty/blocked response
            if "candidates" not in result:
                print(f"   ⚠️  No candidates in response: {result.get('promptFeedback', '')}")
                return None

            text = result["candidates"][0]["content"]["parts"][0]["text"]
            return text if text.strip() else None

        except Exception as e:
            print(f"   ❌ Attempt {attempt+1} failed: {e}")
            time.sleep(10)

    return None

# ── Main loop ─────────────────────────────────────────────
success = 0
failed = 0
skipped = 0
total = len(single_page)

print(f"🔍 Starting OCR with safer rate limiting...\n" + "─"*60)

for i, filename in enumerate(single_page, start=1):
    txt_path = os.path.join(PDF_TEXT_DIR, filename.replace(".pdf", ".txt"))

    if os.path.exists(txt_path):
        skipped += 1
        continue

    try:
        doc = fitz.open(os.path.join(PDF_DIR, filename))
        page = doc[0]
        mat = fitz.Matrix(2, 2)
        pix = page.get_pixmap(matrix=mat)
        image_bytes = pix.tobytes("png")
        doc.close()

        text = gemini_ocr(image_bytes)

        if text and len(text.strip()) > 20:
            with open(txt_path, "w", encoding="utf-8") as f:
                f.write(f"Source PDF: {filename}\n")
                f.write(f"OCR Method: Gemini Vision\n\n")
                f.write(text)
            success += 1
            print(f"[{i}/{total}] ✅ {filename[:55]}")
        else:
            failed += 1
            print(f"[{i}/{total}] ⚠️  Empty: {filename[:55]}")

        # Safer rate limit: 1 request every 6 seconds = 10/min (under 15 limit)
        time.sleep(6)

        # Every 50 files, take a longer break
        if i % 50 == 0:
            print(f"\n😴 Pause 30s to reset rate limit...\n")
            time.sleep(30)

    except Exception as e:
        print(f"[{i}/{total}] ❌ ERROR: {filename[:50]} → {e}")
        failed += 1
        time.sleep(6)

print("─"*60)
print(f"\n📊 OCR Results:")
print(f"   ✅ Success  : {success}")
print(f"   ❌ Failed   : {failed}")
print(f"   ⏭️  Skipped  : {skipped}")
````

````python
import json
import os

# ── Paths ─────────────────────────────────────────────────
SYNTHETIC_DIR = "/content/drive/MyDrive/Colab Notebooks/IUB_Synthetic_QA"
os.makedirs(SYNTHETIC_DIR, exist_ok=True)

# ── 1. Process all_merges_file.json (best file) ───────────
# Upload it to Colab first then run:

with open("/content/all_merges_file.json", "r", encoding="utf-8") as f:
    merged_data = json.load(f)

# Clean and filter
clean_qa = []
seen_answers = set()

SKIP_CATEGORIES = [
    "generic", "weather", "motivation", "robotics",
    "future", "languages", "self_improvement"
]

for item in merged_data:
    q = item.get("question", "").strip()
    a = item.get("answer", "").strip()
    cat = item.get("category", "").strip()

    # Skip empty
    if not q or not a:
        continue

    # Skip generic non-IUB content
    if any(skip in cat.lower() for skip in SKIP_CATEGORIES):
        continue

    # Skip duplicate answers
    a_key = a[:100]
    if a_key in seen_answers:
        continue
    seen_answers.add(a_key)

    # Skip answers that repeat themselves (hallucination sign)
    sentences = a.split(".")
    if len(sentences) > 3:
        if sentences[0].strip() == sentences[1].strip():
            continue

    clean_qa.append({
        "question": q,
        "answer": a,
        "category": cat,
        "source": "kaggle_all_merges"
    })

print(f"✅ all_merges_file.json:")
print(f"   Original entries : {len(merged_data)}")
print(f"   Clean entries    : {len(clean_qa)}")
print(f"   Removed          : {len(merged_data) - len(clean_qa)}")

# Save
with open(os.path.join(SYNTHETIC_DIR, "kaggle_merged_qa.json"),
          "w", encoding="utf-8") as f:
    json.dump(clean_qa, f, ensure_ascii=False, indent=2)

# ── 2. Process intents.json (selective) ───────────────────
with open("/content/intents.json", "r", encoding="utf-8") as f:
    intents_data = json.load(f)

USEFUL_TAGS = [
    "hours", "number", "course", "admission", "scholarship",
    "facilities", "hostel", "fee", "campus", "department",
    "library", "transport", "exam", "result", "contact",
    "IUB", "iub", "university", "Univeristy", "intake",
    "eligib", "merit", "program", "degree"
]

intents_qa = []
for intent in intents_data:
    tag = intent.get("tag", "")
    patterns = intent.get("patterns", [])
    responses = intent.get("responses", [])

    # Only keep IUB-relevant intents
    if not any(kw.lower() in tag.lower() for kw in USEFUL_TAGS):
        continue

    if not responses:
        continue

    best_answer = max(responses, key=len)

    for pattern in patterns:
        if isinstance(pattern, str) and len(pattern) > 5:
            intents_qa.append({
                "question": pattern,
                "answer": best_answer,
                "category": tag,
                "source": "kaggle_intents"
            })

print(f"\n✅ intents.json:")
print(f"   Original intents : {len(intents_data)}")
print(f"   Useful Q&A pairs : {len(intents_qa)}")

# Save
with open(os.path.join(SYNTHETIC_DIR, "kaggle_intents_qa.json"),
          "w", encoding="utf-8") as f:
    json.dump(intents_qa, f, ensure_ascii=False, indent=2)

print(f"\n📦 Total Kaggle Q&A added : {len(clean_qa) + len(intents_qa)}")
print(f"📁 Saved to : {SYNTHETIC_DIR}")
````
